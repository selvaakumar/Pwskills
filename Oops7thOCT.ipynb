```python
#Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance.
#Implement methods to deposit and withdraw money from the account.

class Bank:

  def __init__(self,AccountNo,name,balance):
    self.AccNo = AccountNo
    self.Name = name
    self.__balance = balance

  def Check_balance(self):
    return(self.__balance)

  def withdraw(self,amount):
    if amount < self.__balance and self.__balance > 0:
      self.__balance = self.__balance - amount
      return (f"Amount Withdrawn of {amount}")
    else:
      return("No suffficent balance.")

  def Deposite(self,amount):
      if amount > 0:
        self.__balance = self.__balance + amount
        return (f"Amount Deposited and new balance is {self.__balance}")
      else:
        return ("Enter amount Greater than 0")
```

```python
#Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

class Vehicle:
    def __init__(self, vehicle_id, vehicle_type, is_available=True):
        self.vehicle_id = vehicle_id
        self.vehicle_type = vehicle_type
        self.is_available = is_available


class VehicleRentalSystem:
    def __init__(self):
        self.vehicles = []

    def add_vehicle(self, vehicle_id, vehicle_type):
        vehicle = Vehicle(vehicle_id, vehicle_type)
        self.vehicles.append(vehicle)

    def rent_vehicle(self, vehicle_id):
        for vehicle in self.vehicles:
            if vehicle.vehicle_id == vehicle_id and vehicle.is_available:
                vehicle.is_available = False
                return f"Vehicle {vehicle_id} rented successfully."
        return f"Vehicle {vehicle_id} is not available for rent."

    def return_vehicle(self, vehicle_id):
        for vehicle in self.vehicles:
            if vehicle.vehicle_id == vehicle_id and not vehicle.is_available:
                vehicle.is_available = True
                return f"Vehicle {vehicle_id} returned successfully."
        return f"Vehicle {vehicle_id} is not rented or does not exist."

    def display_available_vehicles(self):
        available_vehicles = [vehicle for vehicle in self.vehicles if vehicle.is_available]
        if available_vehicles:
            print("Available Vehicles:")
            for vehicle in available_vehicles:
                print(f"Vehicle ID: {vehicle.vehicle_id}, Type: {vehicle.vehicle_type}")
        else:
            print("No vehicles available.")


if __name__ == "__main__":
    rental_system = VehicleRentalSystem()

    # Adding vehicles
    rental_system.add_vehicle("V001", "Car")
    rental_system.add_vehicle("V002", "SUV")
    rental_system.add_vehicle("V003", "Truck")

    # Display available vehicles
    rental_system.display_available_vehicles()

    # Rent a vehicle
    print(rental_system.rent_vehicle("V002"))

    # Display available vehicles after renting
    rental_system.display_available_vehicles()

    # Return the rented vehicle
    print(rental_system.return_vehicle("V002"))

    # Display available vehicles after returning
    rental_system.display_available_vehicles()

```

```python
#Problem 4: Library Catalog Create classes representing a library and a book.
#Implement methods to add books to the library, borrow books, and display available books.

class Book:
  def __init__(self,Name,is_available = True):
     self.Name = Name
     self.is_available = is_available


class Library:
  def __init__(self):
     self.books = []

  def add_Books(self,Name):
    book = Book(Name)
    self.books.append(book)
    return f"Book is added to library {Name}"

  def Borrow_book(self,name):
    for books in self.books:
      if books.Name == name and books.is_available == True:
        books.is_available = False
        return "Book Borrowed"
      else:
        return "No such book available in Library"

  def Return_book(self,name):
    for book in self.books:
      if book.Name == name and book.is_available ==False:
        book.is_available = True
        return "Book Returned"
      else:
        return "No such book available in Library"


  def available_books(self):
      available = [book for book in self.books if book.is_available == True]
      print("Available Books")
      for b in available:
        print(str(b.Name))





```

```python
#Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

class Shape:
    def __init__(self, length, width, height):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        """
        Calculate the surface area of the shape.
        """
        area = 2 * (self.length * self.width + self.length * self.height + self.width * self.height)
        return area

    def calculate_perimeter(self):
        perimeter = 4 * (self.length + self.width)
        return perimeter



```

```python
length = 5
width = 3
height = 4

# Create a shape
rectangular_prism = Shape(length, width, height)

# Calculate area and perimeter
area = rectangular_prism.calculate_area()
perimeter = rectangular_prism.calculate_perimeter()

print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
```

```python
#Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.
class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def calculate_average_grade(self):
        """
        Calculate the average grade of the student.
        """
        if not self.grades:
            return 0.0  # Return 0 if there are no grades

        total_grade = sum(self.grades)
        average_grade = total_grade / len(self.grades)
        return average_grade

    def display_student_details(self):
        """
        Display the student details.
        """
        print("Student ID:", self.student_id)
        print("Name:", self.name)
        print("Grades:", self.grades)
        print("Average Grade:", self.calculate_average_grade())

```

```python
student_id = "S12345"
student_name = "John Doe"
student_grades = [85, 90, 92, 88, 78]

# Create a student
student = Student(student_id, student_name, student_grades)

# Display student details and average grade
student.display_student_details()
```

```python
#Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, content):
        self.posts.append(content)

    def display_posts(self):
        print(f"Posts for {self.username}:")
        for index, post in enumerate(self.posts, start=1):
            print(f"Post {index}: {post}")

    def search_posts_by_keyword(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if matching_posts:
            print(f"Posts containing '{keyword}':")
            for index, post in enumerate(matching_posts, start=1):
                print(f"Matching Post {index}: {post}")
        else:
            print(f"No posts found containing '{keyword}'.")
```

```python
username = "john_doe"
profile = SocialMediaProfile(username)

# Add some posts
profile.add_post("Having a great day at the beach! #vacation")
profile.add_post("Enjoying my favorite book. #reading #relaxing")
profile.add_post("Hiking with friends in the mountains. #adventure")

# Display all posts
profile.display_posts()

# Search for posts containing a keyword
keyword_to_search = "beach"
profile.search_posts_by_keyword(keyword_to_search)
```

```python
#Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates.
# Implement methods to add tasks, mark tasks as completed, and display pending tasks.

class ToDo:
  def __init__(self,task,duedate,completed = False):
    self.task = task
    self.duedate = duedate
    self.completed = completed

class Todolist:
  def __init__(self):
    self.lists = []

  def add_todo(self,task,duedate):
    todo = ToDo(task,duedate)
    self.lists.append(todo)
    return "Todo list Added Successfuly"

  def mark_completed(self, task):
    for i in self.lists:
      if i.task == task and i.completed == False:
        i.completed = True
        return "List Marked as Completed."

  def Pending_list(self):
    l = [task for task in self.lists if task.completed == False ]
    for i in l:
      print(f"List Name : {i.task} Due date : {i.duedate}")



```
