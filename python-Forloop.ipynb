12. Create a program that prints the prime numbers in a given range using a `for` loop.

```python
def is_prime(num):
    if num <= 1:
        return False
    elif num <= 3:
        return True
    elif num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return False
        i += 6
    return True

start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))

print(f"Prime numbers in the range [{start}, {end}]:")
for num in range(start, end + 1):
    if is_prime(num):
        print(num, end=' ')
```

20. Write a program to find the greatest common divisor (GCD) of two numbers using a `for` loop.

```python
def gcd(num1, num2):
    gcd_result = 1
    for i in range(1, min(num1, num2) + 1):
        if num1 % i == 0 and num2 % i == 0:
            gcd_result = i
    return gcd_result

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

result = gcd(num1, num2)
print(f"The GCD of {num1} and {num2} is {result}")
```

31. Write a program to generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list
comprehension.

```python
def sieve_of_eratosthenes(n):
    sieve = [True] * (n + 1)
    sieve[0] = sieve[1] = False

    for p in range(2, int(n**0.5) + 1):
        if sieve[p]:
            for i in range(p * p, n + 1, p):
                sieve[i] = False

    primes = [p for p in range(2, n + 1) if sieve[p]]
    return primes

# Input the upper limit 'n' for generating prime numbers
n = int(input("Enter an upper limit (n) for generating prime numbers: "))
prime_list = sieve_of_eratosthenes(n)
print("Prime numbers up to", n, "are:", prime_list)
```

32. Create a program that generates a list of all Pythagorean triplets up to a specified limit using list
comprehension.

```python
def generate_pythagorean_triplets(limit):
    triplets = [(a, b, c) for a in range(1, limit + 1)
                           for b in range(a, limit + 1)
                           for c in range(b, limit + 1)
                           if a**2 + b**2 == c**2]
    return triplets

# Input the limit for generating Pythagorean triplets
limit = int(input("Enter a limit for generating Pythagorean triplets: "))
pythagorean_triplets = generate_pythagorean_triplets(limit)
print("Pythagorean triplets up to", limit, "are:", pythagorean_triplets)
```

34. Write a program that calculates the mean, median, and mode of a list of numbers using list
comprehension.

```python
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    numbers.sort()
    n = len(numbers)
    if n % 2 == 0:
        middle1 = numbers[n // 2]
        middle2 = numbers[n // 2 - 1]
        return (middle1 + middle2) / 2
    else:
        return numbers[n // 2]

def calculate_mode(numbers):
    num_counts = {}
    for num in numbers:
        if num in num_counts:
            num_counts[num] += 1
        else:
            num_counts[num] = 1

    mode = [k for k, v in num_counts.items() if v == max(num_counts.values())]
    if len(mode) == len(numbers):
        return "No unique mode found"
    return mode

# Input a list of numbers
input_str = input("Enter a list of numbers separated by spaces: ")
numbers = [float(x) for x in input_str.split()]

# Calculate mean, median, and mode
mean_result = calculate_mean(numbers)
median_result = calculate_median(numbers)
mode_result = calculate_mode(numbers)

# Print the results
print("Mean:", mean_result)
print("Median:", median_result)
print("Mode:", mode_result)
```

35. Create a program that generates Pascal's triangle up to a specified number of rows using list
comprehension.

```python
def generate_pascals_triangle(rows):
    if rows <= 0:
        return []

    triangle = [[1]]
    for _ in range(1, rows):
        # Calculate the next row using list comprehension
        previous_row = triangle[-1]
        next_row = [1] + [previous_row[i - 1] + previous_row[i] for i in range(1, len(previous_row))] + [1]
        triangle.append(next_row)

    return triangle

# Input the number of rows for Pascal's triangle
num_rows = int(input("Enter the number of rows for Pascal's triangle: "))
pascals_triangle = generate_pascals_triangle(num_rows)

# Print Pascal's triangle
print("Pascal's Triangle with", num_rows, "rows:")
for row in pascals_triangle:
    print(' '.join(map(str, row)).center(2 * num_rows - 1))
```

36. Develop a program that calculates the sum of the digits of a factorial of numbers from 1 to 5 using list
comprehension.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

def sum_of_digits(n):
    return sum(int(digit) for digit in str(n))

numbers = [1, 2, 3, 4, 5]
factorial_sums = [sum_of_digits(factorial(num)) for num in numbers]

for i, num in enumerate(numbers):
    print(f"The sum of digits of {num}! is {factorial_sums[i]}")
```

40. Write a program that generates a list of prime palindromic numbers using list comprehension.

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def is_palindrome(n):
    return str(n) == str(n)[::-1]

prime_palindromes = [n for n in range(1, 1001) if is_prime(n) and is_palindrome(n)]

print("Prime palindromic numbers from 1 to 1000 are:", prime_palindromes)
```
