```python
#18 Write a Python program to determine the roots of a quadratic equation using if-else.

import math

def calculate_quadratic_roots(a, b, c):
    discriminant = b**2 - 4*a*c

    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return f"Roots are real and different: {root1}, {root2}"
    elif discriminant == 0:
        root = -b / (2*a)
        return f"Roots are real and the same: {root}"
    else:
        real_part = -b / (2*a)
        imaginary_part = math.sqrt(abs(discriminant)) / (2*a)
        return f"Roots are complex: {real_part} + {imaginary_part}i, {real_part} - {imaginary_part}i"

def main():
    try:
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))

        if a == 0:
            print("Coefficient 'a' cannot be zero. Not a quadratic equation.")
            return

        roots = calculate_quadratic_roots(a, b, c)
        print("Roots of the quadratic equation:", roots)

    except ValueError:
        print("Invalid input. Please enter valid numerical values.")
main()

```

```python
#5 Write a Python program that calculates the factorial of a number using the `reduce()` function.

def factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    if n == 0 or n == 1:
        return 1
    return reduce(lambda x, y: x * y, range(1, n + 1))


number = 5
result = factorial(number)

print(f"The factorial of {number} is: {result}")
```

```python
#6 Create a Python program that uses `reduce()` to find the GCD (Greatest Common Divisor) of a list of numbers.

from functools import reduce
from math import gcd

def calculate_gcd(a, b):
    """Calculate the GCD of two numbers using the math.gcd() function."""
    return gcd(a, b)

def find_gcd(numbers):
    """Find the GCD of a list of numbers using the reduce() function."""
    if not numbers:
        return None
    return reduce(calculate_gcd, numbers)

numbers = [24, 36, 48, 60]

gcd_result = find_gcd(numbers)

print("List of numbers:", numbers)
print("GCD of the numbers:", gcd_result)
```

```python
#8 Write a Python program that uses the `reduce()` function to find the sum of the digits of a given number.

from functools import reduce

def calculate_digit_sum(total, digit):
    """Calculate the sum of two digits."""
    return total + int(digit)

def sum_of_digits(number):
    """Calculate the sum of digits of a given number using reduce()."""
    if number < 0:
        raise ValueError("Input number must be non-negative")

    digits = str(number)

    digit_sum = reduce(calculate_digit_sum, digits, 0)

    return digit_sum

number = 12345

digit_sum = sum_of_digits(number)

print("Number:", number)
print("Sum of digits:", digit_sum)

```

```python
#3 Create a recursive Python function to find the nth Fibonacci number.
def fibonacci(n):
    """Return the nth Fibonacci number using recursion."""
    if n <= 0:
        raise ValueError("Input must be a positive integer")
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(10)
```

```python
#4 Write a recursive Python function to calculate the sum of all elements in a list.

def recursive_sum(lst):
    """Calculate the sum of all elements in a list using recursion."""
    if not lst:
        return 0
    else:
        return lst[0] + recursive_sum(lst[1:])

numbers = [1, 2, 3, 4, 5]

result = recursive_sum(numbers)

print("List of numbers:", numbers)
print("Sum of all elements:", result)

```

```python
#9 Write a recursive Python function to find all permutations of a given string.


def permutations(string):
    """Generate all permutations of a given string."""
    def generate_permutations(prefix, remaining):
        if len(remaining) == 0:
            result.append(prefix)
        else:
            for i in range(len(remaining)):
                generate_permutations(prefix + remaining[i], remaining[:i] + remaining[i+1:])

    result = []
    generate_permutations("", string)
    return result

input_string = "abc"
perms = permutations(input_string)

print(f"Permutations of '{input_string}':")
for perm in perms:
    print(perm)
```

```python
#10 Write a recursive Python function to check if a string is a palindrome.

def is_palindrome(s):
    """Check if a string is a palindrome using recursion."""
    if len(s) <= 1:
        return True

    if s[0] == s[-1]:
        return is_palindrome(s[1:-1])
    else:
        return False

input_str = "radar"
print(f"Is '{input_str}' a palindrome? {is_palindrome(input_str)}")

input_str = "hello"
print(f"Is '{input_str}' a palindrome? {is_palindrome(input_str)}")
```
